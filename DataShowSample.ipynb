```python
from PIL import Image
import glob, random
import matplotlib.pyplot as plt
import numpy as np 
```

```python
outFile = "SushiPhotos.npz"
maxPhoto = 200
photoSize = 64
xPath = []
yLabel = []
```

```python
# All Images to Numpy Format
def saveImages():

    # Load the Image of a Folder
    readImages("./PhotoData/Bonito", 0)
    readImages("./PhotoData/CaliforniaRoll", 1)
    readImages("./PhotoData/CongerEel", 2)
    readImages("./PhotoData/FattyTuna", 3)
    readImages("./PhotoData/InariSushi", 4)
    readImages("./PhotoData/LeanTuna", 5)
    readImages("./PhotoData/Mackerel", 6)
    readImages("./PhotoData/Salmon", 7)
    readImages("./PhotoData/SalmonRoe", 8)
    readImages("./PhotoData/Scallop", 9)
    readImages("./PhotoData/SeaUrchin", 10)
    readImages("./PhotoData/Shrimp", 11)
    readImages("./PhotoData/SushiBurrito", 12)
    readImages("./PhotoData/Tamago", 13)
    readImages("./PhotoData/TunaRoll", 14)

    # Save to File
    np.savez(outFile, x=xPath, y=yLabel)
    print("Save completed : " + outFile, len(xPath))
```

```python
# Labeling Images in a Path
def readImages(path, label):

    files = glob.glob(path + "/*.jpg")
    random.shuffle(files)
    num = 0
    
    for f in files:
        if num >= maxPhoto: break
        num += 1
        img = Image.open(f)
        img = img.convert("RGB")
        img = img.resize((photoSize, photoSize))
        img = np.asarray(img)
        xPath.append(img)
        yLabel.append(label)
```

```python
# Save the Image Data to Numpy Format
saveImages()
```
